```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_df = pd.read_csv(school_data_to_load)
student_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
combined_df = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])

combined_df = combined_df.sort_values(by = ['school_name'], ascending = True)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Pulling district school data: Total Schools, Students, Money Budgeted.
#Then district student data: Average Math & Reading Scores w/ their Pass Rates.

totalschools = school_df['school_name'].count()

totalstudents = school_df['size'].sum()

totalbudget = "${:,.2f}".format(school_df['budget'].sum())

avemath = combined_df['math_score'].mean()

avereading = combined_df['reading_score'].mean()

#Filtering the dataframe to show passing math students.
math70 = combined_df[combined_df['math_score'] >= 70]

math70percent = '{:.2f}%'.format(((math70['math_score'].count())/totalstudents)*100)

reading70 = combined_df[combined_df['reading_score'] >= 70]

reading70percent = '{:.2f}%'.format(((reading70['reading_score'].count())/totalstudents)*100)

#If they already passed math, then I just have to check if that group also passed reading.
both70 = math70[math70['reading_score']>=70]


both70percent = '{:.2f}%'.format(((both70['math_score'].count())/totalstudents)*100)

#Creating the dictionary for the table.
summarydict = {'Total Schools':[totalschools],'Total Students':[totalstudents],
               'Total Budget':[totalbudget],'Average Math Score':[avemath],
              'Average Reading Score':[avereading], '% Passing Math':[math70percent],
              '% Passing Reading':[reading70percent],'% Overall Passing':[both70percent]}

#Creating the desired dataframe.
summary_df = pd.DataFrame.from_dict(summarydict)

summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Sorted the school data by school name then saving their values: Names, Types, Student Numbers, etc.
sorted_df = school_df.sort_values(by = ['school_name'], ascending = True)

names_sch = sorted_df['school_name']

type_sch = sorted_df['type']

students_sch = sorted_df['size']

budget_sch = sorted_df['budget']

studentbudget_sch = budget_sch / students_sch


#Now we wanted student data of each of the schools, so we first grouped the data by school name.
#Then found the same data from before about math & reading averages and % passing rates.
schoolgroups = combined_df.groupby('school_name')

studentmath = schoolgroups['math_score'].mean().tolist()

studentreading = schoolgroups['reading_score'].mean().tolist()

mathgroups = math70.groupby('school_name')

math70percent_sch = (((mathgroups['math_score'].count().sort_index()).tolist())/students_sch)*100

readinggroups = reading70.groupby('school_name')

reading70percent_sch = (((readinggroups['reading_score'].count().sort_index()).tolist())/students_sch)*100

both70groups = both70.groupby('school_name')

both70percent_sch = (((both70groups['math_score'].count().sort_index()).tolist())/students_sch)*100

#Creating the dictionary for the table.
summarydict_sch = {'School Type': type_sch,'Total Students':students_sch,
               'Total School Budget':budget_sch,'Per Student Budget':studentbudget_sch,
              'Average Math Score': studentmath, 'Average Reading Score': studentreading,
               '% Passing Math':math70percent_sch,'% Passing Reading':reading70percent_sch,
               '% Overall Passing':both70percent_sch}

schoolsummary_df = pd.DataFrame.from_dict(summarydict_sch)

#This dataframe needs the school names to make sense, feed it back in as the index, but remove the column title.
schoolsummary_df.index = names_sch
schoolsummary_df.index.name = None

schoolsummary_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Each time the combined data frame is first filtered by grade
#Then grouped by school names
#Then math scores school averages are calculated.

grade9 = combined_df[combined_df['grade']=='9th']
grade9grp = grade9.groupby('school_name')
grade9math = grade9grp['math_score'].mean()

grade10 = combined_df[combined_df['grade']=='10th']
grade10grp = grade10.groupby('school_name')
grade10math = grade10grp['math_score'].mean()

grade11 = combined_df[combined_df['grade']=='11th']
grade11grp = grade11.groupby('school_name')
grade11math = grade11grp['math_score'].mean()

grade12 = combined_df[combined_df['grade']=='12th']
grade12grp = grade12.groupby('school_name')
grade12math = grade12grp['math_score'].mean()


#Put it all together into a dictionary and a dataframe.
mathgradedict = {'9th':grade9math, '10th': grade10math, '11th': grade11math, '12th': grade12math}

mathgrade_df = pd.DataFrame.from_dict(mathgradedict)

#School names are put back in as a the index but unnecessary column title is removed.
mathgrade_df.index = names_sch
mathgrade_df.index.name = None

mathgrade_df
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
#Exact same idea only the reading score school averages are being calculated.

grade9reading = grade9grp['reading_score'].mean()

grade10reading = grade10grp['reading_score'].mean()

grade11reading = grade11grp['reading_score'].mean()

grade12reading = grade12grp['reading_score'].mean()

readinggradedict = {'9th':grade9reading, '10th': grade10reading, 
                    '11th': grade11reading, '12th': grade12reading}

readinggrade_df = pd.DataFrame.from_dict(readinggradedict)

readinggrade_df.index = names_sch
readinggrade_df.index.name = None

readinggrade_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Bins were created as directed

bins = [0, 585, 630, 645, 680]

bin_names = ["<$585", "$580-630", "$630-645", "$645-680"]

spendingbins = schoolsummary_df.copy()

#Added the cut column to the dataframe.
spendingbins["Spending Ranges (Per Student)"] = pd.cut(spendingbins["Per Student Budget"], bins,right=False, labels=bin_names)

#Grouped by this new spending ranges.
spending_grps = spendingbins.groupby('Spending Ranges (Per Student)')

#The averages for all the indcated bins are calculated.
spending_grps[['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']].mean()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#Almost the same ideas as the previous task, just grouping by sizes instead this time.
sizebins = [0, 1000, 2000, 5000]

sizebin_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

size_df = schoolsummary_df.copy()

size_df["School Size"] = pd.cut(size_df["Total Students"], sizebins,right=False, labels=sizebin_names)

size_grps = size_df.groupby('School Size')

size_grps[['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']].mean()
```
