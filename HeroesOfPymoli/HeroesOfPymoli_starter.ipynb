## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create bins in which to place values based upon TED Talk views
bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]

bin_sizes = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

# Slice the data and place it into bins

herobins = heroes.copy()

herobins['Age Ranges'] =  pd.cut(herobins["Age"], bins, right=False, labels=bin_sizes)

heroes_age = herobins.groupby('Age Ranges')


#Find purchase count, average purchase price, and total purchase values.
age_totals = heroes_age['Gender'].count()

age_priceave = heroes_age['Price'].mean().map("${:.2f}".format)

age_pricetotal = heroes_age['Price'].sum().map("${:.2f}".format)

age_purchaseave = (heroes_age['Price'].sum() / heroes_age['SN'].count()).map("${:.2f}".format)



agedict = {'Purchase Count':age_totals, 'Average Purchase Price': age_priceave, 'Total Purchase Value': age_pricetotal, 'Avg Total Purchase per Person': age_purchaseave}

#Create a dataframe to display the information.
age_df = pd.DataFrame.from_dict(agedict)

age_df.index = bin_sizes
age_df.index.name = "Age Ranges"

age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
usergroups = heroes.groupby('SN')

user_index = usergroups['Purchase ID'].count().index.tolist()

user_purchcount = usergroups['Purchase ID'].count().tolist()

user_purchase = usergroups['Price'].mean().tolist()

userdict = {'Purchase Count': user_purchcount, 'Average Purchase Price':user_purchase}

user_df = pd.DataFrame.from_dict(userdict)

user_df.index = user_index
user_df.index.name = "SN"

user_df['Total Purchase Value'] = (user_df['Purchase Count']*user_df['Average Purchase Price']).map("${:.2f}".format)

user_df['Average Purchase Price'] = user_df['Average Purchase Price'].map("${:.2f}".format)

user_df = user_df.sort_values(by = ['Purchase Count'], ascending= False)

user_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items_df = heroes[['Item ID','Item Name','Price']]

#This would be an alternative method that would do all the calculations at once, I'm not sure how to then clean up the formatting.
#itemgroups = items_df.groupby(['Item ID','Item Name']).agg({'Item Name': ['count'],'Price':['mean','sum']})

itemgroups = items_df.groupby(['Item ID','Item Name'])

items_count = itemgroups['Item Name'].count()

itemaveprice = itemgroups['Price'].mean().map("${:.2f}".format)

itemsumprice = itemgroups['Price'].sum().map("${:.2f}".format)

itemdict = {'Purchase Count': items_count, 'Item Price': itemaveprice, 'Total Purchase Value': itemsumprice}

items_summarydf = pd.DataFrame.from_dict(itemdict)

items_purchasedf = items_summarydf.sort_values(by = ['Purchase Count'], ascending = False)

items_purchasedf
```
